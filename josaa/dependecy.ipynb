```python
from itertools import combinations
import pandas as pd
df = pd.read_csv('totaldata.csv', index_col=0)
x = list(df.columns)
fd_cols = []
subsets = []
for i in range(1, len(x) + 1):
    for i in list(combinations(x, i)):
        subsets.append(list(i))

for i in subsets:
    for j in x:
        if df.groupby(i)[j].nunique().max() == 1:
            fd_cols.append((i, j))

if len(fd_cols) > 0:
    print("Functional dependencies found between the following column pairs:")
    for pair in fd_cols:
        if pair[1] not in pair[0]:
            print(pair[0], "->", pair[1])
else:
    print("No functional dependencies found.")
```
